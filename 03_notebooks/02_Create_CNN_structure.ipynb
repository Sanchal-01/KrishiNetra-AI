```python
import os  # Folder management ke liye zaruri hai
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.layers import Dropout
from tensorflow.keras.preprocessing.image import ImageDataGenerator 
```

```python
train_datagen = ImageDataGenerator(rescale=1./255)

val_datagen = ImageDataGenerator(rescale=1./255)


# SECTION 3: TRAINING GENERATOR
train_generator = train_datagen.flow_from_directory(
    '02_dataset/01_train',
    target_size=(150, 150),
    batch_size = 10,
    class_mode='categorical'
)

# SECTION 4: VALIDATION GENERATOR
val_generator = val_datagen.flow_from_directory(
    '02_dataset/02_validate',
    target_size=(150, 150),
    batch_size= 10,
    class_mode='categorical'
)
```

```python

# 3. Building Our !st CNN Model :

model = models.Sequential()

# C1: Convolution
model.add(layers.Conv2D(filters=32, kernel_size=(3,3), activation='relu', input_shape=(150,150,3)))

# S2: Max Pooling
model.add(layers.MaxPooling2D(pool_size=(2,2)))

# C3: Convolution
model.add(layers.Conv2D(filters=64, kernel_size=(3,3), activation='relu'))

# S4: Max Pooling
model.add(layers.MaxPooling2D(pool_size=(2,2)))

# C5: Convolution
model.add(layers.Conv2D(filters=64, kernel_size=(3,3), activation='relu'))

# S6: Max Pooling
model.add(layers.MaxPooling2D(pool_size=(2,2)))

# Flatten
model.add(layers.Flatten())

# Fully Connected Layers
model.add(layers.Dense(128, activation='relu'))
model.add(Dropout(0.1))
model.add(layers.Dense(64, activation='relu'))
model.add(Dropout(0.1))

# Output Layer
model.add(layers.Dense(19, activation='softmax'))

# 4. Compile Model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])


# Explanation: Ye AI ka "Brain Structure" hai. Conv2D filter ki tarah edges aur colors dhoondta hai.
# MaxPooling sirf zaruri details rakhta hai. 
# Aakhir mein Dense(5) humein 5 plants ke hisaab se result deta hai.


model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# 3. TRAINING

print("\n Training started ........")
history = model.fit(
    train_generator,
    epochs=10,
    validation_data= val_generator
)
```

```python

# 4. SAVE MODEL
# Creating a folder to save our trained 'Brain' (Model).
if not os.path.exists('04_Models'):
    os.makedirs('04_models')

model.save('04_models/plant_model_v1_pro.keras')
print("\n Pro Model save ho gaya")
```
